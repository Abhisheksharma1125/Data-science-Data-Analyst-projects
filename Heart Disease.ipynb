```python
import numpy as np
import pandas as pd
import seaborn as sns

dt=pd.read_csv('Heart.csv',encoding='unicode_escape')
```

```python
dt.shape
```

```python
dt.info()
```

```python
pd.isnull(dt).sum()
```

```python
dt.dropna(inplace=True)
```

```python
dt.columns
```

```python
sns.countplot(x='sex',data=dt)
```

```python
fx=sns.countplot(x='sex',data=dt)

for bars in fx.containers:
    fx.bar_label(bars)
```

```python
fx=sns.countplot(x='age',hue='sex',data=dt)
sns.set(rc={'figure.figsize':(40,8)})

for bar in fx.containers:
    fx.bar_label(bar)
```

```python
fx=dt.groupby(['age','sex'],as_index=False)['chol'].sum().sort_values(by='chol',ascending=True)
sns.set(rc={'figure.figsize':(40,8)})
sns.barplot(data=dt,x='age',y='chol',hue='sex')

```
