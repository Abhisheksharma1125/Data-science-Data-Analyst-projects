```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dt=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
dt.shape
```

```python
dt.info()
```

```python
dt.head()
```

```python
dt.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
dt.info()
```

```python
pd.isnull(dt).sum()
```

```python
dt.dropna(inplace=True)
```

```python
dt.shape
```

```python
dt['Amount']=dt['Amount'].astype('int')
```

```python
dt['Amount'].dtypes
```

```python
dt.columns
```

```python
dt.rename(columns={'Marital_Status':'shaadi'})
```

```python
dt.describe()
```

```python
dt[['Age','Orders','Amount']].describe()
```

```python
sns.countplot(x='Gender',data=dt)
```

```python
fx=sns.countplot(x='Gender',data=dt)

for bars in fx.containers:
    fx.bar_label(bars)
```

```python
dt.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
fx=sns.countplot(data=dt,x='Age Group',hue='Gender')

for bar in fx.containers:
    fx.bar_label(bar)
```

```python
sales_age=dt.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
sales_state=dt.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state_Amount=dt.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state_Amount)
```

```python
fx=sns.countplot(data=dt,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bar in fx.containers:
    fx.bar_label(bar)
```

```python
sales_state=dt.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

```python
fx=sns.countplot(data=dt,x='Occupation')
sns.set(rc={'figure.figsize':(20,8)})
for bar in fx.containers:
    fx.bar_label(bar)
```

```python
sales_state=dt.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')

```

```python
fx=sns.countplot(data=dt,x='Product_Category')
sns.set(rc={'figure.figsize':(25,5)})
for bar in fx.containers:
    fx.bar_label(bar)
```

```python
sales_data=dt.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(33,5)})
sns.barplot(data=sales_data,x='Product_Category',y='Amount')
```
